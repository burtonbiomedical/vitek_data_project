```python
import numpy as np
import pandas as pd
from collections import defaultdict
import math
import matplotlib.pyplot as plt
```

```python
years = list(range(2007, 2018, 1))
mic_values = defaultdict(list)
y = 0
for year in years:
    mic_values[year] = list(map(lambda x: round(x*4)/4,np.random.normal(0.5+y, 0.2, 500).clip(min=0)))
    y += .05
```

```python
mean_mics = dict()
for year, mics in mic_values.items():
    mean_mics[year] = np.mean(mics)
```

```python
sorted_means = sorted(mean_mics.items())
x,y = zip(*sorted_means)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('MIC')
plt.savefig('web_app/public/img/example_figs/mean_mic')
plt.show()
```

```python
mic_df = pd.DataFrame(mic_values)
for col in mic_df.columns:
    mic_df[col].hist(bins=5)
    plt.title(col)
    plt.xlabel('MIC')
    plt.ylabel('Frequency')
    plt.savefig('web_app/public/img/example_figs/hist/{}'.format(col))
    plt.show()
```
