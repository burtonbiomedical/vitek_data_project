<h1>Modify Organism Index collection</h1> 

Correct the organism index collection that the organism name is in a valid key value pair

```python
import pymongo
client = pymongo.MongoClient()
db = client['vitekAlpha']
```

```python
org_indexes = db.organism_indexes
```

```python
obj_list = []
for doc in org_indexes.find():
    org_name = list(filter(lambda x: x != '_id', list(doc.keys())))[0]
    new_index_obj = {
        'organism_name' : org_name,
        'reports': doc[org_name]
    }
    obj_list.append(new_index_obj)
```

```python
obj_list
```

```python
for doc in obj_list:
    db.new_organism_index.insert_one(doc)
```
